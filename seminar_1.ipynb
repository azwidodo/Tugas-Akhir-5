```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
relu = lambda x: max(0, x)
tanh = lambda x: np.tanh(x)
```

```python
x = np.arange(-8, 8, 0.1)
y = [relu(i) for i in x]

plt.rcParams["font.size"] = 14
plt.plot(x, y, "orangered")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("ReLU")
plt.grid(True)
plt.savefig("ReLU.png", bbox_inches="tight")
```
